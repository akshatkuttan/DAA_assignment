```python
import time
import matplotlib.pyplot as plt
import sys
```

```python
sys.setrecursionlimit(10**6)

def iterative_sum(N):
    sum = 0
    for i in range(1, N + 1):
        sum += i
    return sum

def recursive_sum(N, acc=0):
    if N == 0:
        return acc
    else:
        return recursive_sum(N - 1, acc + N)

def measure_time(func, N):
    start_time = time.time()
    func(N)
    end_time = time.time()
    return end_time - start_time

N_values = [10, 100, 1000, 10000, 100000]  # Different values of N to test
iterative_times = []
recursive_times = []

for N in N_values:
    iterative_time = measure_time(iterative_sum, N)
    recursive_time = measure_time(recursive_sum, N)
    iterative_times.append(iterative_time)
    recursive_times.append(recursive_time)
```

```python
plt.plot(N_values, iterative_times, label='Iterative')
plt.plot(N_values, recursive_times, label='Recursive')
plt.xlabel('N')
plt.ylabel('Time (s)')
plt.title('Time taken by Iterative vs Recursive Sum Algorithms')
plt.legend()
plt.show()
```

```python
import time
import random
import matplotlib.pyplot as plt
```

```python
def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

def binary_search(arr, key):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == key:
            return mid
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1
    return -1

arr = [random.randint(1, 1000) for _ in range(10000)]

search_times_linear = []
search_times_binary = []

for _ in range(5):
    key = int(input("Enter the search key: "))
    
    start_time = time.time()
    linear_search(arr, key)
    end_time = time.time()
    search_times_linear.append(end_time - start_time)
    
    arr.sort()  
    start_time = time.time()
    binary_search(arr, key)
    end_time = time.time()
    search_times_binary.append(end_time - start_time)


search_numbers = [i for i in range(1, 6)]

plt.plot(search_numbers, search_times_linear, label='Linear Search')
plt.plot(search_numbers, search_times_binary, label='Binary Search')
plt.xlabel('Search Number')
plt.ylabel('Time (s)')
plt.title('Time taken by Linear vs Binary Search')
plt.legend()
plt.show()
```

```python
def string_to_integer(s):
    # Base case: if the string is empty, return 0
    if not s:
        return 0
    
    # Convert the first character to integer
    digit = int(s[0])
    
    # Recursive call to convert the rest of the string
    remaining = string_to_integer(s[1:])
    
    # Calculate the integer represented by the string
    return digit * (10 ** (len(s) - 1)) + remaining

# Example usage:
string = "13531"
result = string_to_integer(string)
print("String:", string)
print("Integer:", '{:,}'.format(result))
```

```python
def reverse_string(s):
    # Base case: if the string is empty or has only one character, return the string itself
    if len(s) <= 1:
        return s
    else:
        # Recursive call: reverse the substring excluding the first and last characters
        return s[-1] + reverse_string(s[1:-1]) + s[0]

# Example usage:
string = "pots&pans"
reversed_string = reverse_string(string)
print("Original String:", string)
print("Reversed String:", reversed_string)
```

```python
def is_palindrome(s):
    # Base case: if the string is empty or has only one character, it's a palindrome
    if len(s) <= 1:
        return True
    else:
        # Check if the first and last characters are the same
        if s[0] == s[-1]:
            # Recursive call: check if the substring excluding the first and last characters is a palindrome
            return is_palindrome(s[1:-1])
        else:
            return False

# Example usage:
strings = ["racecar", "gohangasalamiimalasagnahog", "hello", "level"]
for string in strings:
    print(f"{string}: {is_palindrome(string)}")
```
